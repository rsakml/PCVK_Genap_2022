```python
# 7.	Buat rectangle dan circle pada bagian wajah dari image foto anda saat beraktifitas (bukan pasfoto).

my_photo = cv.imread('/content/drive/MyDrive/Pengolahan Citra/images/rossa.jpg')

my_photo = cv.cvtColor(my_photo, cv.COLOR_BGR2RGB)

cv.rectangle(my_photo, pt1=(400, 530), pt2=(630, 760), color=(255, 0, 0), thickness=10)

cv.circle(my_photo, center=(510, 650), radius=130, color=(0, 255, 0), thickness=8)

plt.imshow(my_photo)
```

```python
# 8.	Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers! 

kitten = cv.imread('/content/drive/MyDrive/Pengolahan Citra/images/kitten01.jpg')

kitten_rgb = cv.cvtColor(kitten, cv.COLOR_BGR2RGB)

cv.rectangle(kitten_rgb, pt1=(0, 350), pt2=(150, 500), color=(0, 0, 0), thickness=5)

plt.imshow(kitten_rgb)
```
